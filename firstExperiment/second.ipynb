```python
def is_prime(x):
    if x < 2:
        return False
    for i in range(2, x):
        if i * i > x:
            break
        if x % i == 0:
            return False
    return True
```

```python
MIN = 2
MAX = 1000000
while not is_prime(MAX):
    MAX -= 1
print(MIN, MAX)
```

```python
def print_info(a):
    print(a)
    q = 0
    s = 0
    for x in a:
        if not is_prime(x):
            q += 1
        s += x
    s /= len(a)
    d = 0
    for x in a:
        d += (x - s) * (x - s)
    d /= len(a)
    print("wrong tests:",q)
    print("mean:",s)
    print("variance:",d)
    print("Consist min:", MIN in a)
    print("Consist max:", MAX in a)

    
```

```python
print_info(a1)
print_info(a2)
print_info(a3)
```
